```python
import pandas as pd
import plotly
import plotly.express as px
```

```python
churn_data = pd.read_csv('data/churn.csv')
```

```python
churn_df = churn_data.copy()
```

```python
churn_df.info()
```

```python
#  9.1

#    
counts = churn_df['Exited'].value_counts()

# 
fig = px.pie(
    churn_df, # 
    values='CustomerId',
    names='Exited',
    title='Customer ratio' #
)

# 
fig.show()
print(counts)
```

```python
#  9.2
histogram_data = churn_df[churn_df['Balance'] >2500] #   

# 
fig = px.histogram(
    histogram_data, # 
    x='Balance', # 
    labels={'Balance':'   , $'}, #   x
    marginal='box', # 
    nbins=30, # 
)

# 
fig.show()
```
