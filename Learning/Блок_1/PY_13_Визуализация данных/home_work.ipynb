```python
import pandas as pd
import plotly
import plotly.express as px
```

```python
churn_data = pd.read_csv('data/churn.csv')
```

```python
churn_df = churn_data.copy()
```

```python
churn_df.info()
```

```python
# Задание 9.1

#количество лояльных и ушедших клиентов
counts = churn_df['Exited'].value_counts()

#строим график
fig = px.pie(
    churn_df, #табличные данные
    values='CustomerId',
    names='Exited',
    title='Customer ratio' #заголовок
)

#отображаем его
fig.show()
print(counts)
```

Лояльных клиентов почти 80%, ушедших немного больше 20%. Соотношение лояльных и ушедших клиентов 4 к 1.

```python
# Задание 9.2
histogram_data = churn_df[churn_df['Balance'] >2500] #фильтр по размеру баланса

#строим график
fig = px.histogram(
    histogram_data, #табличные данные
    x='Balance', #ось абсцисс
    labels={'Balance':'Баланс на счету клиента, $'}, # название оси x
    marginal='box', #дополнительный график
    nbins=30, #количество интервалов
)

#отображаем график
fig.show()
```

В основном клиенты с балансом более 2500 долларов хранят на счетах от 100 до 140 тысяч долларов. Модальное значение находится в диапазоне 120k-130k долларов и приближено к медиане. Гистограмма напоминает нормальное распределение Гаусса.
