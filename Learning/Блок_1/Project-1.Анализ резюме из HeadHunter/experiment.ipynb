```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import kaleido
```

```python
hh_data = pd.read_csv('data/dst-3.0_16_1_hh_database.csv', sep=';')
course_data = pd.read_csv('data/ExchangeRates.csv', sep=',')
```

```python
hh_df = hh_data.copy()
```

```python


#строим график
fig = px.histogram(
    hh_df, #табличные данные
    x='ЗП', #ось абсцисс
    title='Распределение заработной платы', #заголовок
    labels={'ЗП (руб)':'Заработная плата, руб.'}, #название оси x
    marginal='box', #дополнительный график
    height=500, #высота
    width=800 #ширина
)

#форматирование
fig.update_layout(
    margin=dict(l=20, r=20, t=60, b=20),
    paper_bgcolor='LightSteelBlue'
)

#отображаем график
fig.show()
#записываем в файл
fig.write_html("plotly/chart3_1.html")
```
