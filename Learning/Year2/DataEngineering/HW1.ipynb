```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import numpy as np 
```

```python
driver_path = 'chromedriver'

def run(driver_path):
    driver = webdriver.Chrome(
        service=ChromeService(executable_path=driver_path)
    )
    # Указываем сайт 
    url = 'https://tproger.ru/quiz/real-programmer/'
    driver.get(url)

    # Устанавливаем задержку, чтобы сайт успел прогрузиться
    driver.implicitly_wait(5)

    # Фиксируем название теста 
    paragraph = driver.find_element(By.CLASS_NAME, 'tp-content-paragraph').text
    print(paragraph, '\n')

    # Фиксируем количество вопросов
    count_answer = np.arange(0, 12)

    # С помощью цикла перебираем страницы
    for x in count_answer:
        # Фиксируем вопрос
        question = driver.find_element(By.CLASS_NAME, 'tp-quiz-question__text').text
        print(question, '\n')

        # Фиксируем в списке все элементы по классу, и с помощью цикла отображаем варианты ответов
        answers = driver.find_elements(By.CLASS_NAME, 'tp-quiz-answer-list__list')
        for x in answers:
            print(x.text)

        click_answer = driver.find_element(By.XPATH, '//*[@id="quiz-app"]/div/div/div/div[3]/ul/li[1]')
        click_answer.click()
        driver.implicitly_wait(5)

        true_answer = driver.find_element(By.CLASS_NAME, 'tp-answer--result-true').text
        print('Правильный ответ: ', true_answer)

        click_next = driver.find_element(By.CLASS_NAME, 'tp-button--primary')
        driver.execute_script('arguments[0].click()', click_next)
        driver.implicitly_wait(5)

        print('*' * 80)
    driver.quit()
```

```python
run(driver_path)
```
