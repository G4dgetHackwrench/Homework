```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import tree
from sklearn import model_selection
from sklearn import metrics
import plotly.express as px
```

```python
voice_data = pd.read_csv('data/voice_gender.csv')
voice_data.head()
```

```python
voice_data.info()
```

```python
voice_data.isnull().sum().sum()
```

```python
X = voice_data.drop('label', axis=1)
y = voice_data.label

# Формируем обучающую и тестовую выборки
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print('Train shape: {}'.format(X_train.shape))
print('Test shape: {}'.format(X_test.shape))
```

```python
dt = tree.DecisionTreeClassifier(
    max_depth=1,
    criterion='entropy'
)
dt.fit(X_train, y_train)

tree.plot_tree(
    decision_tree=dt, #дерево решений 
    feature_names=X.columns, #имена факторов
    filled=True, #расцветка
    impurity=True, #отображать ли неоднородность в вершинах
);
```

```python
1296/2534
```

```python
print(round(1296/2534*100, 1))
y_pred = dt.predict(X_test)
print(round(metrics.accuracy_score(y_test, y_pred), 3))
```

```python
dt = tree.DecisionTreeClassifier(
    max_depth=2,
    criterion='entropy'
)
dt.fit(X_train, y_train)

tree.plot_tree(
    decision_tree=dt, #дерево решений 
    feature_names=X.columns, #имена факторов
    filled=True, #расцветка
    impurity=True, #отображать ли неоднородность в вершинах
    class_names=dt.classes_
);
```

```python
y_pred = dt.predict(X_test)
print(round(metrics.accuracy_score(y_test, y_pred), 3))
```

```python
dt = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
dt.fit(X_train, y_train)

tree.plot_tree(
    decision_tree=dt, #дерево решений 
    feature_names=X.columns, #имена факторов
    filled=True, #расцветка
    impurity=True, #отображать ли неоднородность в вершинах
    class_names=dt.classes_
);
```

```python
#глубина дерева
print(dt.get_depth())
#количество листьев
print(dt.get_n_leaves())
y_pred_train = dt.predict(X_train)
y_pred = dt.predict(X_test)
print(round(metrics.accuracy_score(y_train, y_pred_train), 3))
print(round(metrics.accuracy_score(y_test, y_pred), 3))
```

```python
# Задаём сетку параметров
param_grid = {
    'criterion': ['gini', 'entropy'], #критерий информативности
    'max_depth': [4, 5, 6, 7, 8, 9, 10], #максимальная глубина дерева
    'min_samples_split': [3, 4, 5, 10] #минимальное количество объектов, необходимое для сплита
}

# Задаём метод кросс-валидации
cv = model_selection.StratifiedKFold(n_splits=5)

grid_search = model_selection.GridSearchCV(
    estimator=tree.DecisionTreeClassifier(criterion='entropy', random_state=0), 
    param_grid=param_grid, 
    cv=cv,
    scoring='accuracy',
    n_jobs = -1
) 

#Обучаем модель
%time grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
#считаем предсказания с наилучшими гиперпараметрами
y_pred_train = grid_search.predict(X_train)
y_pred = grid_search.predict(X_test)

print(round(metrics.accuracy_score(y_train, y_pred_train), 3))
print(round(metrics.accuracy_score(y_test, y_pred), 3))
```

```python
dt = tree.DecisionTreeClassifier(
    criterion='gini', #критерий информативности
    max_depth=7,
    min_samples_split=3,
    random_state=0
)
dt.fit(X, y)
```

```python
data = pd.DataFrame({'Name':X_train.columns, 'Ratings':dt.feature_importances_})
data = data.sort_values('Ratings')
fig = px.bar(data,
             x='Ratings',
             y='Name',
             height=400, width=800,
             text_auto=True,
             orientation='h'
)

fig.update_layout(
    margin=dict(l=20, r=20, t=60, b=20),
    paper_bgcolor='LightSteelBlue',
    showlegend=False
)

fig.show('png')
```
